# Bingo boards generator

Takes as input a text file of bingo squares with one prompt per line and outputs randomized bingo boards using pdfLaTex

```python
template = '''\\documentclass{article}
\\usepackage[utf8]{inputenc}
\\usepackage[margin=1in]{geometry}
\\usepackage{cmbright}
\\usepackage{array}
\\newcolumntype{P}[1]{>{\\centering\\arraybackslash}p{#1}}
\\newcommand{\\cc}[1]{\\begin{tabular}{l} #1 \\end{tabular}}

\\title{ESP Retreat Bingo}
\\author{}
\\date{Fall 2022}

\\begin{document}

\\maketitle

\\vspace{2em}

\\begin{center}
\\renewcommand{\\arraystretch}{4}
\\begin{tabular}{|P{2.5cm}|P{2.5cm}|P{2.5cm}|P{2.5cm}|P{2.5cm}|}
\\hline 
PROMPT & PROMPT & PROMPT & PROMPT & PROMPT \\\\
\\hline 
PROMPT & PROMPT & PROMPT & PROMPT & PROMPT \\\\
\\hline 
PROMPT & PROMPT & FREE & PROMPT & PROMPT \\\\
\\hline 
PROMPT & PROMPT & PROMPT & PROMPT & PROMPT \\\\
\\hline 
PROMPT & PROMPT & PROMPT & PROMPT & PROMPT \\\\
\\hline 
\\end{tabular}
\\renewcommand{\\arraystretch}{1}
\\end{center}

\\end{document}
'''
```

```python
f = open('espingo.txt','r')
prompts = f.read().strip().split('\n')
temp = template.split('PROMPT')
f.close()
```

```python
from numpy import random
```

```python
for j in range(60):
    result = ''
    random.shuffle(prompts)
    for i in range(len(temp)-1):
        result += temp[i] + prompts[i]
    result += temp[-1]
    fn = 'output/b'+str(j)+'.tex'
    f = open(fn,'w')
    f.write(result)
    f.close()
    ! pdflatex -output-directory output $fn
```
